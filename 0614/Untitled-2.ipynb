```python
import os
import cv2

def make_square_image(input_path, output_path):
    # 이미지 불러오기
    image = cv2.imread(input_path)

    # 이미지의 너비와 높이 가져오기
    height, width, _ = image.shape

    # 짧은 길이 계산
    min_dim = min(height, width)

    # 짧은 길이를 기준으로 이미지를 정사각형으로 자르기
    cropped_image = image[:min_dim, :min_dim]

    # 정사각형 이미지 저장
    cv2.imwrite(output_path, cropped_image)

# 이미지가 있는 폴더 경로 설정
input_folder = "./download_image_data/"
output_folder = "./cropped_image_data/"

# 폴더 내의 이미지 파일들에 대해 작업 수행
for filename in os.listdir(input_folder):
    file_path = os.path.join(input_folder, filename)
    if os.path.isfile(file_path):
        # 이미지 파일인 경우에만 작업 수행
        image_extension = os.path.splitext(filename)[1].lower()
        if image_extension in ['.jpg', '.jpeg', '.png', '.bmp']:
            # 이미지를 정사각형으로 변환하여 저장할 경로 지정
            output_filename = "square_" + filename
            output_path = os.path.join(output_folder, output_filename)
            
            # 이미지를 정사각형으로 변환하여 저장
            make_square_image(file_path, output_path)
            
            print(f"{filename}을 정사각형으로 변환하여 {output_path}에 저장하였습니다.")
```
