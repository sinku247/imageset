`````python
# 미지 정제 : 이미지 폴더에 있는 파일 가져오기(os.walk)

def get_img_paths(root_path):
    file_paths = []
    for (path, dir, files) in os.walk(root_path):
        print('path',path)
        print('dir',dir)
        print('files',files)
        for file in files:
            ext = os.path.splitext(file[-1].lower())
            print('ext',ext)
            formats = ['bmp', 'png', 'jpg', 'jpeg','.tif', '.tiff','.dng']
            if ext in formats:
                file_path = os.path.join(path, file)
                print('file_path ->',file_path)
                file_paths.append(file_path)

    return file_paths


file_paths = get_img_paths('./download_image_data/')
print(file_paths)

for i in file_paths:
    print(i)
`````
